```python
stats = np.array([[1,2,3],[4,5,6]])
```

```python
stats.sum() # нахождение суммы
```

```python
stats.sum(axis=0) # сумма по столбцу
```

```python
stats.sum(axis=1) # сумма по строкам
```

```python
stats.min(axis=0) # минимальное по колонкам
```

```python
stats.max(axis=1) # максимальное по строкам
```

```python
stats.mean() # среднее в массиве
```
